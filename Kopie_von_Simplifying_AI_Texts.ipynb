```python
from transformers import pipeline
#model belirtmelisin question ve context öncesi, yoksa cvb göremiyorsun

qa_pipeline = pipeline(
    "question-answering",
    model="distilbert/distilbert-base-cased-distilled-squad",
    revision="564e9b5",  # optional
    device=-1  # CPU (using) )(zaten uyarıda belirtilmiş)
)

result = qa_pipeline({
    "context": "The wafer singulation process consists of several steps intended to separate each die by cutting down the middle of the street or scribe lane created between chips",
    "question": "What is the wafer singulation process?"
})

print(result)
```
